```python
import cv2, os
```

```python
class_name = ['Anda', 'Apa', 'Berhenti', 'Bodoh', 'Cantik', 'Halo', 'Hati-hati', 'Lelah', 'Maaf', 'Makan', 'Mau-Ingin', 'Membaca', 'Nama', 'Sama-sama', 'Saya', 'Siapa', 'Sombong', 'Takut', 'Terima kasih']

DATA_PATH = "./terbisa.v3i.yolov11"
```

```python
train_path = os.path.join(DATA_PATH, "train")
test_path = os.path.join(DATA_PATH, "test")
valid_path = os.path.join(DATA_PATH, "valid")
train_images = []
train_labels = []
train_boxes = []

train_image_folder = os.path.join(train_path, "images")
train_label_folder = os.path.join(train_path, "labels")

for image_file, label_file in zip(os.listdir(train_image_folder), os.listdir(train_label_folder)):
    image_path = os.path.join(train_image_folder, image_file)
    label_path = os.path.join(train_label_folder, label_file)

    if image_file[:-4] != label_file[:-4]:
        print("Error: image and label file mismatch")
        continue

    image = cv2.imread(image_path)
    with open(label_path, 'r') as file:
        content = file.read()

    content = content.split()
    try:        
        label = int(content[0])
        # x = float(content[1])
        # y = float(content[2])
        # w = float(content[3])
        # h = float(content[4])
        train_images.append(image)
        train_labels.append(label)
        # train_boxes.append([x, y, w, h])
    except:
        print("Error: label file is empty")
        continue


```

```python
valid_image_folder = os.path.join(valid_path, "images")
valid_label_folder = os.path.join(valid_path, "labels")

for image_file, label_file in zip(os.listdir(valid_image_folder), os.listdir(valid_label_folder)):
    image_path = os.path.join(valid_image_folder, image_file)
    label_path = os.path.join(valid_label_folder, label_file)

    if image_file[:-4] != label_file[:-4]:
        print("Error: image and label file mismatch")
        continue

    image = cv2.imread(image_path)
    with open(label_path, 'r') as file:
        content = file.read()

    content = content.split()
    try:        
        label = int(content[0])
        # x = float(content[1])
        # y = float(content[2])
        # w = float(content[3])
        # h = float(content[4])
        mirrored = cv2.flip(image, 1)
        train_images.append(image)
        train_labels.append(label)
        train_images.append(mirrored)
        train_labels.append(label)

        # train_boxes.append([x, y, w, h])
    except:
        print("Error: label file is empty")
        continue
```

```python
test_images = []
test_labels = []
test_boxes = []

test_image_folder = os.path.join(test_path, "images")
test_label_folder = os.path.join(test_path, "labels")

for image_file, label_file in zip(os.listdir(test_image_folder), os.listdir(test_label_folder)):
    image_path = os.path.join(test_image_folder, image_file)
    label_path = os.path.join(test_label_folder, label_file)

    if image_file[:-4] != label_file[:-4]:
        print("Error: image and label file mismatch")
        continue

    image = cv2.imread(image_path)
    with open(label_path, 'r') as file:
        content = file.read()

    content = content.split()
    # print(image_file[:-4], content)
    try:        
        label = int(content[0])
        # x = float(content[1])
        # y = float(content[2])
        # w = float(content[3])
        # h = float(content[4])
        mirrored = cv2.flip(image, 1)
        test_images.append(image)
        test_labels.append(label)
        test_images.append(mirrored)
        test_labels.append(label)
        # test_boxes.append([x, y, w, h])
    except:
        print("Error: label file is empty")
        continue
```

```python
# COMBINED DATASET
NEWDATA_PATH = "new_dataset"

def save_image():
    if not os.path.exists(NEWDATA_PATH):
        os.makedirs(NEWDATA_PATH)
    if not os.path.exists(os.path.join(NEWDATA_PATH, "combined")):
        os.makedirs(os.path.join(NEWDATA_PATH, "combined"))

    train_folder = os.path.join(NEWDATA_PATH, "combined")
    hash = dict()
    for (image, label) in zip(train_images, train_labels):
        image_label = class_name[label]
        if image_label not in hash:
            hash[image_label] = 0
            os.makedirs(os.path.join(train_folder, image_label))
        image_name = f"{image_label}-{hash[image_label]}.jpg"
        hash[image_label] += 1
        cv2.imwrite(os.path.join(train_folder, image_label, image_name), image)

    for (image, label) in zip(test_images, test_labels):
        image_label = class_name[label]

        image_name = f"{image_label}-{hash[image_label]}.jpg"
        hash[image_label] += 1
        cv2.imwrite(os.path.join(train_folder, image_label, image_name), image)
save_image()
```
